# Joint Plots with positive relationships

### CRIM versus AGE

```python
sns.jointplot(x = 'CRIM', y ='AGE', data = Boston_DataFrame)
sns.lmplot(x = 'CRIM', y = 'AGE', data = Boston_DataFrame, hue = 'CHAS', fit_reg = False)
```

##### Only subhurbs that have a certain number or older houses have higher crime rates.
##### All new subhurbs have low crime rate.
##### Older subhurbs are closer to the river.

### INDUS versus NOX, AGE, TAX and LSTAT

```python
sns.jointplot (x = Boston_DataFrame['INDUS'], y = Boston_DataFrame['NOX'])
sns.lmplot(x = 'INDUS', y = 'NOX', data = Boston_DataFrame, hue = 'CHAS', fit_reg = False)
sns.jointplot (x = Boston_DataFrame['INDUS'], y = Boston_DataFrame['AGE'])
sns.lmplot(x = 'INDUS', y = 'AGE', data = Boston_DataFrame, hue = 'CHAS', fit_reg = False)
sns.jointplot (x = Boston_DataFrame['INDUS'], y = Boston_DataFrame['TAX'])
sns.lmplot(x = 'INDUS', y = 'TAX', data = Boston_DataFrame, hue = 'CHAS', fit_reg = False)
sns.jointplot(x = Boston_DataFrame['INDUS'], y = Boston_DataFrame['LSTAT'])
sns.lmplot(x = 'INDUS', y = 'LSTAT', data = Boston_DataFrame, hue = 'CHAS', fit_reg = False)
```

##### Subhurbs in Boston with more non-retail business have higher NOX concentrations.
##### No newer subhurbs are highly occupied by non-retail businesses.
##### Subhurbs with non-retail business property have higher property taxes.
##### Higher socioeconomic status subhurbs have lower occupation of non-retail businesses.
